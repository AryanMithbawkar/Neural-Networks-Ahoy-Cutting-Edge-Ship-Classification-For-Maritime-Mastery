Description: Build an ANN model for Drug classification.
This project aims to analyze the relationship between various medical parameters and drug effectiveness.
The dataset consists of patient information, including age, sex, blood pressure levels (BP), cholesterol
levels, sodium-to-potassium ratio (Na_to_K), drug type, and corresponding labels. The goal is to develop
a model that can accurately predict the class or category of a given drug based on its features.

```python
#1: Read the dataset and do data pre-processing
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv('drug200.csv')
```

```python
data.head()
```

```python
data.info()
```

```python
data.isnull().value_counts()
```

```python
data['Drug'].value_counts()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
```

```python
data['Sex']=le.fit_transform(data['Sex'])
data['BP']=le.fit_transform(data['BP'])
data['Cholesterol']=le.fit_transform(data['Cholesterol'])
data['Drug']=le.fit_transform(data['Drug'])
```

```python
data.head()
```

```python
data.info()
```

```python
x=data.iloc[:,:5].values
x
```

```python
y=data.iloc[:,5:6].values
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.fit_transform(x_test)
```

```python
x_train.shape,x_test.shape
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
#2: Build the ANN model with (input layer, min 3 hidden layers & output layer)
model=Sequential()
model.add(Dense(5,activation='relu',kernel_initializer='random_uniform'))#input layer
model.add(Dense(10,activation='relu',kernel_initializer='random_uniform'))#1st hidden layer
model.add(Dense(20,activation='relu',kernel_initializer='random_uniform'))#2nd hidden layer
model.add(Dense(15,activation='relu',kernel_initializer='random_uniform'))#3rd hidden layer
model.add(Dense(5,activation='softmax',kernel_initializer='random_uniform'))#output layer
```

```python
model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

```python
model.fit(x_train,y_train,batch_size=32,epochs=300)
```

```python
y_pred=model.predict(x_test)
```
