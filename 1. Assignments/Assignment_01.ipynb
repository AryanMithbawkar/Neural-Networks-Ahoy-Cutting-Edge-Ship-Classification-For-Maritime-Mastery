```python
#1. Write a python Program to display the last digit of number(hint: Any number divide by 10 will return last digit)
a=int(input("enter 2 digit number: "))
b=a%10
print(f"the last digit of number {a} is : {b}")
```

```python
#2. Write a function to Calculate Area and Perimeter of Rectangle
def area(h,b):
  are=h*b
  peri=2*(h+b)
  print(f"area of rectangle: {are}")
  print(f"perimeter of rectangle: {peri}")
c=int(input("enter height of rectangle: "))
d=int(input("enter breadth of rectangle: "))
area(c,d)
```

```python
#3. Write a python Program to find out given No is Prime or Not.
def evo(a):
  if a==0:
    print("number is zero")
  elif a%2==0:
    print(f"{a} is even")
  else:
    print(f"{a} is odd")

x=int(input("enter num:"))
evo(x)
```

```python
#4. Write a Python program that prompts the user to input three numbers and then determines and prints the largest among them.
e=int(input("enter 1st number: "))
f=int(input("enter 2nd number: "))
g=int(input("enter 3rd number: "))
if (e==f and e==g and f==g):
  print(f"all numbers are equal")
elif (e>=f and e>=g):
  print(f"{e} is the largest among all number")
elif (f>=e and f>=g):
  print(f"{f} is the largest among all number")
elif (g>=e and g>=f):
  print(f"{g} is the largest among all number")
```

```python
# 5. Write a Python program to calculate electricity bill (accept no of unit from user) according to following criteria: Unit Price
# First 100 units  no charge
# Next 100 units Rs 5 per Unit
# After 200 Units Rs 10 per Unit
# (For Example if input unit is 350 then total bill amount is 2000)

u=int(input("enter the number of units consumed: "))
if (u<=100):
  print(f"As no. of units consumed is less than or equal to 100 your electricity bill is waiver")
elif (u<=200):
  bill=(u-100)*5
  print(f"As no. of units consumed greater than 100 and less than 200 you will be charged Rs. 5 per unit after free 100 unit")
  print(f"so your electricity bill is: Rs.{bill}")
elif (u>200):
  bill=(100*5)+((u-200)*10)
  print(f"As no. of units consumed greater 200 you will be charged Rs. 5 per unit after free 100 unit and Rs. 10 per unit after 200")
  print(f"so your electricity bill is: Rs.{bill}")
```

```python
#6. Write a program to accept numbers from 1-7 and display name of day like- 1 for sunday, 2 for Monday and so on.
i=int(input("enter number 1-7: "))
match i:
  case 1:
    print(f"Sunday")
  case 2:
    print(f"Monday")
  case 3:
    print(f"Tuesday")
  case 4:
    print(f"Wednesday")
  case 5:
    print(f"Thrusday")
  case 6:
    print(f"Friday")
  case 7:
    print(f"Saturday")
  case _:
    print(f"not a day in week")
```

```python
#7. Write a python program to check given word is Palindrome or not
p=input("enter word to check if palindrome or not: ")
t=p
r=t[::-1]
if(t==r):
  print(f"the word {t} is palindrome word")
else:
  print(f"the word {t} is not palindrome word")
```
